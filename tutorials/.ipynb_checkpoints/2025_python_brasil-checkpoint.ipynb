## Agile Testing Quadrants

![Agile Testing Quadrants](../images/agile_quadrants.png "Os quadrantes do Agile Testing")

Os testes dos quadrantes 1 e 2 s√£o escritos para ajudar o time a entregar um produto que gere valor para os clientes.

### Q1
> "Measures the internal quality of the code", Kent Beck

S√£o testes que fornecem **feedback r√°pido** e s√£o executados antes do c√≥digo ser *merged* na branch principal (*main*).

### Q2
D√£o suporte ao time, garantindo que as funcionalidades estejam alinhadas ao neg√≥cio. Os testes desse quadrante surgem de exemplos fornecidos pelo time de neg√≥cio.

### Q3
S√£o testes feitos do ponto de vista do neg√≥cio, avaliando se o software funciona conforme o especificado ou at√© mesmo comparando com a concorr√™ncia. Nesses testes, simula-se o uso real do usu√°rio. Exemplos incluem testes explorat√≥rios e de usabilidade.

### Q4
S√£o testes que utilizam diversas ferramentas para analisar o produto, discutidos em termos t√©cnicos (e n√£o de neg√≥cio). Envolvem testes de performance, seguran√ßa, carga etc.

## CI/CD

Uma **pipeline de CI/CD** (Integra√ß√£o Cont√≠nua/Entrega Cont√≠nua) √© um conjunto automatizado de etapas que permite que times de desenvolvimento construam, testem e entreguem software de forma r√°pida, segura e frequente.

- **CI (Continuous Integration / Integra√ß√£o Cont√≠nua):**  
  Refere-se ao processo de integrar mudan√ßas de c√≥digo feitas por diferentes pessoas em um reposit√≥rio central, de forma frequente (v√°rias vezes ao dia). - **CD (Continuous Delivery / Entrega Cont√≠nua):**  
  Depois que a integra√ß√£o e os testes passam, o software √© preparado para ser entregue em produ√ß√£o automaticamente. O objetivo √© que qualquer altera√ß√£o aprovada possa ser implantada rapidamente e com seguran√ßa, reduzindo erros e tempo de entrega.

### Exemplo de fluxo de uma pipeline de CI/CD:
1. O desenvolvedor faz um commit/push no reposit√≥rio.
2. A pipeline √© acionada automaticamente.
3. C√≥digo √© compilado/empacotado.
4. Testes automatizados s√£o executados (unit√°rios, integra√ß√£o, E2E).
5. (Opcional) O software √© implantado em ambiente de teste (stage).
6. Se tudo passar, o deploy para produ√ß√£o pode ser feito automaticamente ou com um clique.

**Benef√≠cios:**  
- Detecta problemas cedo.
- Reduz o risco de bugs em produ√ß√£o.
- Torna o processo de deploy mais r√°pido e confi√°vel.
- Permite feedback cont√≠nuo para o time.

## O que s√£o Fixtures no Pytest?

Uma **fixture** no Pytest √© um recurso que permite criar fun√ß√µes para preparar, fornecer e descartar objetos ou estados necess√°rios para os testes. Elas ajudam a evitar repeti√ß√£o de c√≥digo, facilitam o setup e o teardown de ambientes, e tornam os testes mais organizados e reutiliz√°veis.

### Para que servem as fixtures?
- Criar dados de teste ou conex√µes com banco de dados.
- Configurar ambientes (ex: abrir arquivos, iniciar servidores, criar usu√°rios).
- Compartilhar objetos entre m√∫ltiplos testes.
- Garantir que cada teste comece em um estado previs√≠vel e isolado.

### Como funciona?
Voc√™ define uma fun√ß√£o com o decorator `@pytest.fixture` e, nos testes que precisam usar essa fixture, basta adicion√°-la como argumento. O Pytest detecta e executa essa fun√ß√£o antes do teste rodar.

### Exemplo B√°sico

```python
import pytest

@pytest.fixture
def usuario_admin():
    return {"nome": "admin", "senha": "1234"}

def test_login_usuario_admin(usuario_admin):
    assert usuario_admin["nome"] == "admin"
```

Neste exemplo:
- `usuario_admin` √© uma fixture.
- O teste `test_login_usuario_admin` recebe automaticamente o dicion√°rio retornado pela fixture.
